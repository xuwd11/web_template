```python
import json
from copy import deepcopy

def combine_ipynb(file_names, merged_name='merged.ipynb'):
    file_name = file_names[0]
    
    with open(file_names[0], 'r') as f:
        merged = deepcopy(json.load(f))
        
    for file_name in file_names[1:]:
        with open(file_name, 'r') as f:
            merged['cells'] = merged['cells'] + json.load(f)['cells']
    
    with open(merged_name ,'w') as f:
        json.dump(merged, f)

def remove_codes(file_name, new_name=None):
    if new_name is None:
        new_name = file_name[:-6] + '_new.ipynb'
    
    with open(file_name, 'r') as f:
        data = deepcopy(json.load(f))
    
    inds_del = []
    
    for i, cell in enumerate(data['cells']):
        if cell['cell_type'] == 'code':
            inds_del.append(i)
    
    for i, ind_del in enumerate(inds_del):
        del data['cells'][ind_del - i]
        
    with open(new_name, 'w') as f:
        json.dump(data, f)
        
def output2md(file_name, new_name=None, remove_codes=True):    
    if new_name is None:
        new_name = file_name[:-6] + '_new.ipynb'
    
    with open(file_name, 'r') as f:
        data = deepcopy(json.load(f))
        
    cells = []
    
    for cell in data['cells']:
        if cell['cell_type'] != 'code':
            cells.append(deepcopy(cell))
        else:
            if not remove_codes:
                cells.append(deepcopy(cell))
            for output in cell['outputs']:
                if 'data' in output:
                    key = [key for key in output['data'].keys() \
                           if not key in ['text/plain', 'metadata', 'output_type']][0]
                    cells.append({'cell_type': 'markdown', 'metadata':{}, \
                                  'source': deepcopy(output['data'][key])})
                    if 'text/markdown' in output['data']:
                        cells.append({'cell_type': 'markdown', 'metadata':{}, \
                                      'source': deepcopy(output['data']['text/markdown'])})
                    elif 'text/html' in output['data']:
                        cells.append({'cell_type': 'markdown', 'metadata':{}, \
                                      'source': deepcopy(output['data']['text/html'])})
                    
    data['cells'] = cells
    
    with open(new_name, 'w') as f:
        json.dump(data, f)
```

```python
data_dir = '../notebooks/'

file_names = ['index.ipynb', '03_EDA.ipynb', '04_collaborative_filtering.ipynb', \
              '05_content_filtering.ipynb', '06_ensemble.ipynb', \
              '07_results.ipynb', '08_conclusions.ipynb', '09_reference.ipynb']

file_names = [data_dir + file_name for file_name in file_names]
merged_name = data_dir + 'merged.ipynb'

combine_ipynb(file_names, merged_name)
```

```python
data_dir = '../notebooks/'

file_names = ['merged.ipynb']

file_names = [data_dir + file_name for file_name in file_names]

for file_name in file_names:
    output2md(file_name, remove_codes=True)
```
